```python
                           ############ GENERATING STOCHASTIC DATA --> GBM ############

    
import matplotlib.pyplot as plt
import numpy as np

T = 270
mu = 0.002885601
sigma = 0.029523375
S0 = 24.2 # .   ,    
dt = 1 # -:  ""    -  - 
count_experiment = 100 # - ()

def  GBM_generator(mu, sigma, T, dt, S0):
    N = round(T / dt) # -    
    t = np.linspace(0, T, N) #   : , ,   - 
    W = np.random.standard_normal(size=N)
    W = np.cumsum(W) * np.sqrt(dt)  ### standard brownian motion ### #   2000  ,
                                                                     #      -
    X = (mu - 0.5 * sigma ** 2) * t + sigma * W
    S = S0 * np.exp(X)  ### geometric brownian motion ###

    return t, S

for i in range(0, count_experiment):
    gbm_t, gbm_S  = GBM_generator(mu, sigma, T, dt, S0)
    print((gbm_S[-1]))
#     print(np.mean(gbm_S))
    plt.plot(gbm_t, gbm_S, linewidth=1)
```
